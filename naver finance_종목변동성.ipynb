```python
code = {'하나제약':'293480', '피델릭스':'032580', '크라운해태홀딩스':'005740', '퍼시스':'016800', '큐캐피탈':'016600',
       '인텔리인테크':'189300', '진양화학':'051630', '평화정공':'043370', '지투알':'035000', '파세코':'037070',
       '피제이메탈':'128660', '하이텍팜':'106190', '파인디엔씨':'049120', '성광벤드':'014620', '티엘아이':'062860',
       '특수건설':'026150', '하이트론':'019490', '토박스코리아':'215480'}
```

```python
t1 = time.time()

sector_variationlist = []

for i in list(code.values()):
    stockCode = i
    variationlist = []

    dayPriceUrl = 'http://finance.naver.com/item/sise_day.nhn?code=' + stockCode
    dayPriceHtml = urlopen(dayPriceUrl)
    dayPriceSource = BeautifulSoup(dayPriceHtml.read(), "html.parser")

    dayPricePageNavigation = dayPriceSource.find_all("table", align="center")
    dayPriceMaxPageSection = dayPricePageNavigation[0].find_all("td", class_="pgRR")
    #dayPriceMaxPageNum = int(dayPriceMaxPageSection[0].a.get('href')[-3:])

    #for page in range(1, dayPriceMaxPageNum + 1) 하면 나와있는 끝까지

    for page in range(1, 100):
        url = 'http://finance.naver.com/item/sise_day.nhn?code=' + stockCode + '&page=' + str(page)
        html = urlopen(url)
        source = BeautifulSoup(html.read(), "html.parser")
        srlists = source.find_all("tr")
        isCheckNone = None

        # day: 날짜
        # closingPrice: 종가
        # variation: 전일대비
        # openingPrice: 시가
        # highestPrice: 고가
        # lowestPrice: 저가
        # volume: 거래량

        for i in range(1, len(srlists)-1):  
            if (srlists[i].span != isCheckNone):
                           
                variation = srlists[i].find_all('td', class_='num')[1].text.strip()
                openingPrice = srlists[i].find_all("td", class_="num")[2].text
                
                openingPrice2 = openingPrice.replace(',', '')
                variation2 = variation.replace(',', '')
                
                if float(openingPrice2) == 0:
                    pass
                else:
                    percent = float(variation2)/float(openingPrice2) * 100
                    variationlist.append(percent)


    a = np.array(variationlist)  
    sector_variationlist.append(a.mean())
     

t2 = time.time()

print('걸린 시간:', t2-t1)
print(sector_variationlist)
```
