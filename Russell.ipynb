```python
error_ticker_list = []
download_ticker_list = []

def downloading(start_index, stop_index):
    num = 0
    for i in no_have[start_index : stop_index]:
        a = crawler(i)
        try:
            a.file_download()
            download_ticker_list.append(a.ticker)
            num += 1
            print(num)
            
        except Exception as e:
            print('error:', e)
            print('company:', a.ticker)
            error_ticker_list.append(a.ticker)
            driver.refresh()
            num += 1
            print(num)
```
